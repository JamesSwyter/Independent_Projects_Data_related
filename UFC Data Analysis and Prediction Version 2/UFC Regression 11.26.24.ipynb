```python
import pandas as pd
import numpy as np

data = pd.read_csv('Downloads/ufc-master.csv')

pd.set_option('display.max_rows', None)
pd.reset_option('display.max_rows')

data = data.drop(columns=['RWFlyweightRank', 'RWFeatherweightRank', 'RWStrawweightRank','RWBantamweightRank', 'RHeavyweightRank','RLightHeavyweightRank','RMiddleweightRank',
                          'RWelterweightRank','RLightweightRank','RFeatherweightRank','RBantamweightRank', 'RFlyweightRank','RPFPRank','BWFlyweightRank','BWFeatherweightRank',
                          'BWStrawweightRank','BWBantamweightRank','BHeavyweightRank','BLightHeavyweightRank','BMiddleweightRank','BWelterweightRank','BLightweightRank','BFeatherweightRank',
                          'BBantamweightRank','BFlyweightRank','BPFPRank','RedExpectedValue','BlueExpectedValue','Date','Country','Location','TitleBout','WeightClass','Gender',
                          'NumberOfRounds','BMatchWCRank','RMatchWCRank','Finish','FinishDetails','FinishRound','FinishRoundTime',
                          'TotalFightTimeSecs','RedDecOdds','BlueDecOdds','RSubOdds','BSubOdds','RKOOdds','BKOOdds','EmptyArena'])

data = data.dropna()

data['Winner'].replace(['Red','Blue'],[1,0], inplace=True)
data['BetterRank'].replace(['Red','Blue','neither'],[1,-1,0], inplace=True)

data = pd.get_dummies(data, columns=['BlueStance','RedStance'])

#data.count()
#data.info()
data.head()
```

```python
data.info()
```

```python
data = data.drop(columns=['RedFighter','BlueFighter'])
data.info()
```

```python
data['Red Over 34'] = np.where(data['RedAge'] > 34, 1, 0)
data['Blue Over 34'] = np.where(data['BlueAge'] > 34, 1, 0)
data.info()

Winner = data['Winner']
```

```python
data = data.iloc[:,49:]

BlueStance_Other = data['BlueStance_Open Stance'] + data['BlueStance_Southpaw'] + data['BlueStance_Switch']
RedStance_Other = data['RedStance_Open Stance'] + data['RedStance_Southpaw'] + data['RedStance_Switch']

print(BlueStance_Other.value_counts())
print(RedStance_Other.value_counts())
```

```python
data['BlueStance_Other'] = BlueStance_Other
data['RedStance_Other'] = RedStance_Other
data = data.drop(columns=['BlueStance_Open Stance', 'BlueStance_Southpaw', 'BlueStance_Switch', 'RedStance_Open Stance','RedStance_Southpaw',
                  'RedStance_Switch'])
data.info()
```

```python
print(data.BlueStance_Orthodox.value_counts())
print(data.RedStance_Orthodox.value_counts())
print(data.BlueStance_Other.value_counts())
print(data.RedStance_Other.value_counts())
```

```python
from sklearn.model_selection import train_test_split

Variables_Train, Variables_Test, Winner_Train, Winner_Test = train_test_split(data, Winner)
```

```python
print(Variables_Train.head())
print(Variables_Test.head())
print(Winner_Train.head())
print(Winner_Test.head())
```

```python
import statsmodels.api as sm

Variables_Train_Array = np.asarray(Variables_Train)
Winner_Train_Array = np.asarray(Winner_Train)

logit_model=sm.Logit(Winner_Train_Array, Variables_Train_Array)
result = logit_model.fit()
print(result.summary())
```

```python
#Choosing only the most statistically significant variables, and eliminating some collinearity (win streak and longest win streak)

Simplified_Variables_Train = Variables_Train[['WinStreakDif','SigStrDif','AvgTDDif','BetterRank','Red Over 34','Blue Over 34']]
Simplified_Variables_Train_Array = np.asarray(Simplified_Variables_Train)
```

```python
#Still retains lots of predictive power, same as original results

logit_model=sm.Logit(Winner_Train_Array, Simplified_Variables_Train_Array)
result = logit_model.fit()
print(result.summary())
```

```python
#Simplified_Variables_Test = Variables_Test[['WinStreakDif','SigStrDif','AvgTDDif','BetterRank','Red Over 34','Blue Over 34']]
#Simplified_Variables_Test_Array = np.asarray(Simplified_Variables_Test)

Winner_Test_Array = np.asarray(Winner_Test)

Simplified_Variables_Test = Variables_Test[['WinStreakDif','SigStrDif','AvgTDDif','BetterRank','Red Over 34','Blue Over 34']]
Simplified_Variables_Test_Array = np.asarray(Simplified_Variables_Test)

predictions = result.predict(Simplified_Variables_Test_Array)
predictions_binary = list(map(round, predictions)) 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

print(confusion_matrix(Winner_Test_Array, predictions_binary))
print(accuracy_score(Winner_Test_Array, predictions_binary))
```
